### 1.1. Oque são genes?

O gene é um segmento de uma sequência específica de molécula de DNA que responsável pelas características herdadas. Cada gene contém um código (instruções) para produzir uma proteína que desempenha uma função específica no corpo. Cada célula humana tem cerca de 25.000 genes.

</center>

As células humanas contêm 23 pares de cromossomos que é uma longa cadeia de DNA contendo um par de cromossomos sexual (XX nas mulheres ou XY nos homens), mais 22 pares de cromossomos não sexuais denominados autossomos. Os espermatozoides e os óvulos contêm apenas metade deles, ou seja, 23 cromossomos e são transmitidos dos pais para filhos. Isto explica porque as crianças se parecem com seus pais, e/ou podem desenvolver determinadas doenças hereditárias.

Uma célula usa seus genes seletivamente, ou seja, pode ou ativar ou desativar os genes que precisa no momento. Todas as células do corpo (exceto óvulo e espermatozoide) contêm os mesmos genes. Uma celula se torna especializada quando ativa e desativa certos genes para se tornar uma celula muscular ou óssea por exemplo, e alguns genes ficam ativos o tempo todo para produzir proteinas basicas para o corpo. ([Oncoguia](http://www.oncoguia.org.br/conteudo/o-que-sao-genes/8159/73/), 2015)

#### 1.1.1. Genes ligados ao Cromossomo X

Normalmente, temos 2 cromossomos sexuais, as mulheres dois cromossomos X, enquanto os homens têm um cromossomo X e um Y. Se o cromossomo Y contém diferentes genes que o cromossomo X, os homens têm apenas uma cópia dos genes do cromossomo X. Algumas doenças/condições são causadas por genes no cromossomo X. No caso do daltonismo, uma mulher tem que ter 2 cópias do gene (um em cada cromossomo X) para ter a condição. No caso do homem, ele só tem que ter o gene no seu cromossomo X. Doenças e condições como estas são denominadas de ligadas ao cromossomo X e são mais frequentes em homens.   A expressão genética ou expressão gênica é o processo pelo qual a informação hereditária contida em um gene, tal como a sequência de DNA, é utilizada de modo a formar um produto génico funcional, tal como proteínas ou RNA. Apesar da importância do entendimento de somente um gene, estudos em redes são mais interessantes em casos práticos.

  Os genes, podem especificar polipeptídios (proteínas) ou RNAs funcionais, traduzindo-os  em estruturas que estão presentes em um determinado tipo celular. Durante a expressão de um gene codificante de proteína, a informação flui do DNA→ RNA→ PROTEÍNA. Esse fluxo direcional de informação é conhecido como o dogma central da biologia molecular, e o processo de ir de um DNA para um produto funcional, é conhecido como expressão gênica.

  Graças a expressão gênica, é possível ter uma diferenciação celular. A célula consegue controlar as proteínas que produz por alguns fatores, por exemplo, controlando a transcrição, a saida e a degradação proteina da celula. Esses fatores, controlam a expressão gênica, aumentando ou diminuindo a sintese de proteínas. ([Wikipédia](https://pt.wikipedia.org/wiki/Express%C3%A3o_g%C3%A9nica))

### 1.3. Mecanismo de Ativação (MOA)

Em farmacologia, a expressão mecanismo de ação refere-se a interação bioquímica específica através da qual uma droga produz um efeito. Um mecanismo de ação usualmente inclui menção de um "alvo" molecular específico no qual a droga liga-se, tal como uma enzima ou receptor.

As enzimas possuem mecanismos de ação que possuem tratamento específico na bioquímica, nos processos biológicos e relacionados, e não é tal conceito diretamente relacionado com o que seja mecanismo de ação da farmacologia.

Por exemplo, o mecanismo de ação da aspirina envolvendo a inibição irreversível da enzima ciclooxigenase, a qual suprime a produção de prostaglandinas e tromboxanos consequentemente reduzindo a dor e inflamação.

Observe-se que qualquer substância ingerida ou administrada sob qualquer forma, possuirá um mecanismo de ação, mesmo em sua toxidade, que pode ir de uma normalmente consumida até em nossa alimentação, como o etanol (bebidas alcoólicas) até as mais venenosas, embora o termo, claramente, seja mais usado em relação a fármacos.

#### 1.3.1. Agonista x Antagonista

Antagonismo em farmacologia se refere aos compostos químicos que se ligam a determinados receptores neurológicos porém sem ativá-los, impedindo que os componentes que o ativariam de se ligarem. O oposto do antagonista é o agonista.

Os antagonistas são muito usados para diminuir o efeito de compostos prejudiciais. Se um paciente tem excesso de serotonina existem vários medicamentos que vão se ligar a um dos tipos de receptores da serotonina sem ativá-los, impedindo que a própria serotonina de se ligar e os ativar, diminuindo o efeito desse tipo de receptor no organismo.

Agonistas ativam receptores que produzem a resposta desejada. Os agonistas convencionais aumentam a proporção de receptores ativados. Os agonistas inversos estabilizam o receptor na sua conformação inativa e agem de forma semelhante aos antagonistas competitivos.

```python
# Removendo as colunas que não podem ir para o modelo para não vazar dados.
colunas_to_drop = ['id', 'droga', 'eh_antagonista',
                   'soma_agonista', 'soma_antagonista'] + \
                    nomes_agonista + nomes_antagonista
x = dados_tratados.drop(columns=colunas_to_drop, axis=1)
x
```

```python
# Predizendo com a variavel mais presente.
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
# Rodando uma lista de modelos e vendo suas pontuações.
lista_ml = [RandomForestClassifier, DecisionTreeClassifier, 
            LinearSVC, SVC, KNeighborsClassifier]
nomes_ml = ['RF', 'DT', 'LSVC', 'SVC', 'KNC']

treinar_modelo_classificador(lista_ml, nomes_ml)
```

O modelo é levemente melhor do que um classificador burro, enquanto que o DecisionTreeClassifier consegue ser pior que o dummy...

#### 4.3.1. Rodando Auto ML

```python
# AutoML, com 50 iterações e com configuração mais leve.
tpot = TPOTClassifier(generations=5, population_size=10, 
                     verbosity=2, random_state=42, config_dict='TPOT light')
tpot.fit(x_treino, y_treino)
print(tpot.score(x_teste, y_teste))
tpot.export('tpot_pipeline.py')
```

```python
# Rodando novamente com os parâmetros obtidos pelo autoMl.
DTtpot = DecisionTreeClassifier(criterion="entropy", 
                                max_depth=2, min_samples_leaf=12, 
                                min_samples_split=7)
DTtpot.fit(x_treino, y_treino)
predict = DTtpot.predict(x_teste)

# Visualizando a pontuação do modelo com um relatorio:
score = classification_report(y_teste, predict)
print(score)
```

Ao rodar um AutoMl, se espera obter uma resultado significativamente maior do que os outros modelos, que por sua vez estão com os hiperparâmetros padrão. Entretanto o tpot melhorou e muito a Árvore de Decisão, que tem uma vantagem que é a sua explicabilidade, podendo gerar um gráfico em formato de árvore.
E atráves do relatorio de classificação, podemos notar que prever se é antagonista é mais fácil, é possisel que isso se deva a ter mais variaveis dessa categoria, entretanto ainda sim é pouca a diferença.
